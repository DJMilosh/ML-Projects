## Laplace-Beltrami operator

Neka je $M$ kompaktna Riemannova mnogostrukost s metrikom $g$. Tražimo preslikavanje $f:M\rightarrow \mathbb{R}$ koje šalje bliske točke na $M$ u bliske točke na $\mathbb{R}$ (u određenom preciznom smislu kojeg želimo motivirati). Pretpostavimo da je $f$ dvaput diferencijabilna.

Za početak, primijetimo da za $x,y\in M$ koje su međusobno blizu imamo $|f(x)-f(y)|=d(x,y)|\nabla f(x)|+o(d(x,y))$, gdje je $d(x,y)$ Riemannova udaljenost na $M$ (infimum duljina svih krivulja koje vežu $x$ i $y$), a $\nabla f(x) \in T_xM$ je dual od $df_x$ (s obzirom na metriku Riemannovu $g$, tj. $df=g(\nabla f, \cdot)$).

Neka je $l=d(x,y)$ i $\gamma$ geodezik (parametriziran duljinom) na kojem se postiže udaljenost, onda $|\gamma'(t)|=1$ u svim $t$. Prema Stokesovom teoremu:

$$
f(y)-f(x)=\int_0^ldf(\gamma'(t))dt=\int_0^lg( \nabla f(\gamma(t)), \gamma'(t) ) dt
$$
S druge strane, Cauchy-Schwarzova nejednakost daje $|g(\nabla f(\gamma(t)), \gamma'(t))|\leq |\nabla f(\gamma(t))| |\gamma'(t)|=|\nabla f(\gamma(t))|,$
a Taylorov teorem $|\nabla f(\gamma(t))|=|\nabla f(x)|+ O(t)$. Iz ovoga iščitavamo:

$$
|f(x)-f(y)|\leq \int_0^l |\nabla f(x)|dt +\int_0^l O(t)dt=l|\nabla f(x)| +o(l),
$$
što je i trebalo pokazati.

Dakle, preslikavanje koje najbolje čuva lokalnost (u nekom globalnom prosjeku) ćemo dobiti tako da minimiziramo tzv. Dirichletovu energiju $\int_M |\nabla f|^2 dV$ po svim $f$ sa $||f||_{L^2}=1$ (inače uzimanjem $f$ sa dovoljno malom $L^2$ normom možemo integral učiniti koliko hoćemo malenim).

Ako s $\mathcal{L}_X$ označimo Liejevu derivaciju duž vektorskog polja $X$, onda prema Stokesu imamo:
$$
0=\int_{\partial M} i_x(f dV)=\int_Mdi_X(fdV)
$$
jer je $M$ bez granice, tj. $\partial M = \emptyset$. Alternativno (ako $M$ ima granicu), ovo vrijedi za svaku $f$ koja na $\partial M$ iščezava (čiji je nosač sadržan u interijeru   mnogostrukosti). Od sada nadalje pretpostavljamo ili da $\partial M =\emptyset$ ili da smo se ograničili na prostor funkcija koje na $\partial M$ iščezavaju.

Prisjetimo se Cartanove formule $\mathcal{L}_X=i_Xd+di_X$ i da je $fdV$ forma najvišeg reda (top forma), čija vanjska derivacija iščezava $d(fdV)=0$. Ovo nam garantira da $\mathcal{L}_X(fdV)=di_X(fdV)$. Dalje, divergencija polja $X$ je određena relacijom $\mathcal{L}_XdV=\nabla \cdot X dV$, a za funkcije $\mathcal{L}_Xf=df(X)$. Konačno, koristeći pravilo produkta za Liejevu derivaciju, imamo:
$$
0=\int_M \mathcal{L}_X (fdV)=\int_M (\mathcal{L}_Xf)dV+ \int_Mf \mathcal{L}_XdV=\int_M df(X)dV+\int_M \nabla \cdot X f dV.
$$
Drugim riječima, za svako vektorsko polje $X$ imamo:
$$
\int_M \langle X, \nabla f \rangle dV=- \int_M (\nabla \cdot X) f dV
$$
pa posebno za $X=\nabla f$ imamo:
$$
\int_M|\nabla f|^2dV=\int_M(-\nabla \cdot \nabla f)f dV
$$

Definiramo li (ipak držeći na umu sukob notacije sa Liejevom derivacijom) tzv. **Laplace-Beltramijev operator** $\mathcal{L}f=-\Delta f=-\nabla^2f= - \nabla \cdot \nabla f$, onda vrijedi:

$$
\langle \mathcal{L}f \vert f \rangle = \int_M (\mathcal{L}f)f dV=\int_M |\nabla f|^2 dV,
$$
gdje je $\langle \cdot \vert \cdot \rangle$ $L^2$ skalarni umnožak.
Dakle, minimiziranjem Dirichletove energije zapravo minimiziramo $\langle \mathcal{L}f \vert f \rangle$ po jediničnim $f$ (u $L^2$ smislu). Kao što smo već nekoliko puta vidjeli, iz ovoga nužno slijedi da $f\in L^2(M)$ mora biti svojstveni vektor od $\mathcal{L}$.

Kako je ovdje riječ o operatorima na beskonačnodimenzionalnim prostorima, možda bi bilo dobro uvjeriti čitatelja da se ne javljaju nikakve značajne komplikacije. Neka je A simetrični operator na nekom Hilbertovom prostoru. Pokušavamo ekstremizirati $\langle v \vert  Av \rangle$ pod uvjetom $|v|=1$. Definirajmo stoga lagranžijan $L=\langle v \vert Av\rangle -\lambda (|v|^2-1)$ te ga derivirajmo (Fréchetovom ili Gauteauxovom derivacijom), što konačno daje $2Av-\lambda2v$. Ovaj posljednji izraz je $0$ upravo onda kada je $v$ svojstveni vektor operatora $A$.



Bitno je napomenuti da ovaj argument ne garantira *postojanje* svojstvenog vektora (postojanje ekstrema). Zaista, u konačnodimenzionalnom slučaju ekstrem na vektorima duljine $1$ (jediničnoj sferi) postoji jer je jedinična sfera kompaktan skup, a $\langle v \vert  Av \rangle$ je neprekidno preslikavanje. U beskonačnodimenzionalnom slučaju jedinična sfera više nije kompaktna, ali se svejedno može pokazati (koristeći npr. Banach-Alougluvljev teorem - vidi https://en.wikipedia.org/wiki/Compact_operator_on_Hilbert_space) da za *kompaktne operatore* $\langle v \vert  Av \rangle$ poprima maksimum na jediničnoj sferi.

Primijetimo da je Laplace-Beltrami simetrični operator. Iz
$$
\int_M \langle X, \nabla f \rangle dV=- \int_M (\nabla \cdot X) f dV
$$
za $X=\nabla g$ imamo
$$
\int_M \langle \nabla g, \nabla f \rangle dV=- \int_M (\nabla \cdot \nabla g) f dV=\int_M(\mathcal{L}g)f dV=\langle \mathcal{L}g \vert f\rangle
$$
Kako je lijeva strana simetrična na zamijenu $f$ i $g$, tako i desna mora biti.

Može se pokazati (vidi *Riemannian
Geometry and Geometric Analysis*, Jürgen Jost ili *Manifolds and Differential Geometry* Jeffrey M. Lee, poglavlje 9) da je spektar operatora $\Delta= - \mathcal{L}$ na kompaktnoj mnogostrukosti diskretan i svaki svojstveni prostor $V_{\lambda}=\{f \ \vert \ \Delta f = \lambda f\}$ je konačnodimenzionalan. Ovo slijedi npr. ako pokažemo da je rezolventa operatora $\Delta$ kompaktni operator na $L^2$ (sami $\Delta$ ne može biti kompaktan jer nije ni omeđen jer kao diferencijalni operator ne može biti definiran na čitavom $L^2(M)$ prostoru). Primjerice, na sferi se svojstvene funkcije Laplace-Beltrami operatora zovu sferni harmonici i bitni su u mnogim granama fizike (u kvantnoj fizici opisuju npr. orbitale vodikovog atoma).

$\mathcal{L}$ je očito pozitivno definitan jer $\langle \mathcal{L} f \vert f\rangle = \int_M |\nabla f|^2 dV\geq 0$. Također, najniža svojstvena vrijednost $\lambda=0$ se postiže i to samo u konstantnim funkcijama - na kompaktnoj (povezanoj) mnogostrukosti (bez granice) su samo konstantne funkcije harmoničke.

Naime, za neprekidnu funkciju $f\geq 0$ imamo $\int_MfdV=0 \implies f=0$ jer bi u suprotnom u nekoj točki $f>0$, ali onda (po neprekidnosti) isto mora vrijediti i u čitavoj okolini $U$ te točke pa bi na toj okolini $\int_UfdV>0$, što je apsurdno ($\int_Uf dV\leq \int_M fdV$ jer je $U\subset M$). Sada odmah imamo da
$$\mathcal{L}f=0\implies \langle \mathcal{L}f \vert f \rangle =0 \implies \int_M |\nabla f|^2 dV=0 \implies |\nabla f|^2=0 \implies f= \text{const}$$

Zapravo, $|\nabla f|=0$ garantira samo da je $f$ lokalno konstantna (oko svake točke postoji okolina na kojoj je $f$ konstantna).  Naravno, ako je $M$ povezan, onda je svaka lokalno konstantna funkcija konstantna pa je u općem slučaju $f$ konstantna na svakoj komponenti. Dakle, iz jezgre Laplace-Beltrami operatora (koja se praktički svede na 0-tu de Rhamovu kohomologiju) se da iščitati broj komponenti.

Označimo svojstvene vrijednosti Laplace-Beltramijevog operatora s $\lambda_i$ i numerirajmo tako da tvore rastući niz $0=\lambda_0<\lambda_1<\dots$. Zapravo, korisnije je svojstvene vektore indeksirati prirodnim brojevima $f_0, f_1, f_2,...$, a onda pustiti da $\lambda_i$ označava svojstvenu vrijednost svojstvenog vektora $f_i$. Sada imamo $0=\lambda_0<\lambda_1\leq \lambda_2 \leq \dots$. Napomenimo da su svojstvene vrijednosti konačnih multipliciteta pa možemo imati samo konačan niz jednakosti $\lambda_k=\dots =\lambda_{k+n}$ prije stroge nejednakosti. Također, kako se $\mathcal{L}f=0$ postiže samo u konstantnim funkcijama, svojstveni prostor pridružen $\lambda=0$ je jednodimenzionalan pa je jedino $\lambda_0=0$ (tj. jedino je $f_0$ konstanta).

Sada je najbolje $d$-dim smještenje (ono koje najbolje čuva lokalnost) dano s $x\mapsto (f_1(x),...,f_{d}(x))$, gdje smo preskočili triviijalni slučaj konstantne funkcije $f_0$.

## Heat kernel

Kažemo da funkcija $u=u(x,t)$, tj. $u:M\times \mathbb{R}\rightarrow \mathbb{R}$ rješava jednadžbu topline ako $(\frac{\partial}{\partial t}+\mathcal{L})u=0$. Kažemo da je $f$ početna distribucija topline ako $u(x,0)=f(x)$. Za početnu distribuciju $f$, rješenje je dano sa $u(x)=\int_MH_t(x,y)f(y)dV(y)$
za neku funkciju $H_t$ koju zovemo **heat kernelom**. Notacija $dV(y)$ govori da treba integrirati po $y$ (tj. integriramo funkciju $y\mapsto H(x,y)f(y)$).

Heat kernel treba zadovoljavati
$$\frac{\partial}{\partial t}H_t(x,y)=\Delta H_t(x,y), \quad \lim_{t\to 0}H_t(x,y)=\delta(x-y)=\delta_x(y),$$
gdje je $\delta$ Diracova delta funkcija (tj. distribucija u smislu teorije Laurenta Schwarza i limes je naravno u distribucijskom smislu). Drugim riječima, početna distribucija topline za heat kernel je Diracov delta, tj. šiljak u x.

U $\mathbb{R}^d$ heat kernel je dan poznatim izrazom $$H_t(x,y)=\frac{1}{(4\pi t)^{d/2}}e^{-|x-y|^2/4t}$$

Na mnogostrukosti $M$ u normalnim koordinatama (gdje lokalno produžimo bazu u nekom tangentnom prostoru $T_pM$ geodezicima) pak imamo
$$
H_t(x,y)=\frac{1}{(4\pi t)^{d/2}}e^{-|x-y|^2/4t}(\phi(x,y)+O(t)),
$$
gdje je $\phi$ glatka i $\phi(x,x)=1$ tako da kada su $x$ i $y$ blizu, a t je malen $H_t(x,y)\approx\frac{1}{(4\pi t)^{d/2}}e^{-|x-y|^2/4t}$ pa približno vrijedi euklidski rezultat.

Za dokaz egzistencije heat kernela i gornji asimptotski razvoj vidi npr. poglavlje 3 u *The Laplacian on a Riemannian manifold: an introduction to analysis on manifolds*, Steven Rosenberg.

Dakle, imamo $\mathcal{L}f(x)=\mathcal{L}u(x,0)=-\frac{\partial}{\partial t}u(x,0)=-\frac{\partial}{\partial t}\vert_{t=0}\int_M H_t(x,y)f(y)dV(y)$. Nastavimo li sa sređivanjem ovog izraza,
$$
\mathcal{L}f(x)=-\lim_{t\to 0}\frac{1}{t}\left( \int_M H_t(x,y)f(y)dV(y)-\int_M H_0(x,y)f(y)dV(y)\right).
$$
S druge strane  (kako $H_t$ teže u Diracov delta),
$$\int_M H_0(x,y)f(y)dV(y)=\lim_{s\to 0}\int_M H_s(x,y)f(y)dV(y)=f(x).$$
Stoga,
$$
\mathcal{L}f(x)= \lim_{t\to 0} \frac{1}{t} \left( f(x)-\int_M H_t(x,y)f(y)dV(y) \right) =\lim_{t\to 0}\frac{1}{t}\left( f(x)\int_MH_t(x,y)dV(y)-\int_MH_t(x,y)f(y)dV(y) \right).
$$
Uvrstimo li izraz za $H_t$ (u normalnim koordinatama):

$$
\mathcal{L}f(x)=\lim_{t\to 0}\frac{1}{t}(4\pi t)^{-d/2}\left( \int_Me^{-|x-y|^2/4t}f(x)dV-\int_Me^{-|x-y|^2/4t}f(y)dV \right).
$$
Naime, $(4\pi t)^{-d/2}e^{-|x-y|^2/4t}$ trne sve više kako je $t$ manji pa u limesu imamo samo euklidski izraz za $H_t$ (daleki $x$ i $y$ malo doprinose, tj. $(4\pi t)^{-d/2}e^{-|x-y|^2/4t}\phi(x,y)\approx(4\pi t)^{-d/2}e^{-|x-y|^2/4t}$).

## Laplacian grafa

Podatci nam leže na mnogostrukosti i tvore graf. Ideja Laplacian eigenmaps algoritma je da Laplace-Beltrami operator mnogostrukosti aproksimira laplasijanom tog grafa. Što je graf, tj. uzorkovanje točaka gušće, to očekujemo da je ta aproksimacija bolja.

*Graf* je konačna kolekcija vrhova $V=\{x_1,...,x_n\}$ te rubova $E\subset V\times V$ i njihovih težina $w_{ij}$ (za rub $(x_i,x_j)$).

Funkcije na vrhovima $f:V\rightarrow \mathbb{R}$ možemo shvatiti kao vektorre u $\mathbb{R}^n$. Naime, $f=(f_1,...,f_n)=(f(x_1),...,f(x_n))$. Prostor funkcija na $V$ tvori (konačnodimenzionalni) Hilbertov prostor sa skalarnim umnoškom $\langle f\vert g\rangle=\sum_i f_ig_i=f^Tg$, kojeg označavamo s $L^2(V)$.

(Nenormalizirani) laplasijan grafa je preslikavanje $L:L^2(V) \rightarrow L^2(V), \quad (Lf)_i=\sum_jw_{ij}(f_i-f_j)$, koje predstavlja razliku između $f$ i njegovog usrednjenja (jer $(\sum_jw_{ij})f_i-\sum_jw_{ij}f_j$). Vidimo da $L$ možemo zapisati i kao $L=D-W$, gdje je $W$ matrica težina $w_{ij}$, a $D$ dijagonalna matrica sa $D_{ii}=\sum_jw_{ij}$.

U praksi sume često ograničavamo na samo $k$ najbližih susjeda na grafu (tj. proglasimo $w_{ij}=0$ za točke koje nisu najbliži susjedi).

Definiramo i Dirichletovu energiju $\frac{1}{2}\sum_{ij}(f_i-f_j)^2w_{ij}=f^TLf=\langle f \vert Lf \rangle$. Ove relacije nije teško pokazati. Naime, primijetimo da
$$\sum_{ij}(f_i-f_j)^2w_{ij}=\sum_{ij}(f_i^2+f_j^2-2f_if_j)w_{ij}=\sum_if_i^2D_{ii}+\sum_jf_jD_{jj}-2\sum_{ij}f_if_jw_{ij}=2f^TDf-2f^TWf=2f^TLf.$$

Analogno problemu minimizacije za Laplace-Beltrami operator, imamo i problem minimizacije za laplasijan grafa.

Minimiziranjem $f^TLf$ pod uvjetom $f^TDf=1$ dobijemo $0=\frac{\partial}{\partial f_i}(f^TLf-\lambda(f^TDf-1))$, tj. $2Lf-2\lambda Df=0\implies Lf=\lambda Df$.

Definiramo li normalizirani laplasijan $\overline{L}=D^{-1}L=I-D^{-1}W$, onda treba riješiti svojstveni problem $\overline{L}f=\lambda f$.

## Konzistentnost

Pokazat ćemo da u određenom smislu normalizirani laplasijan grafa (sa ispravno odabranim težinama) konvergira u Laplace-Beltramijev operator.

Pretpostavimo da uzorkujemo po *uniformnoj distribuciji* (!!) točke $x_1,...,x_n\in M$, koje onda tvore graf. Neka su težine tog grafa $w^t_{ij}=e^{-|x-y|^2/4t}$ (ovisi o $t$). Onda za $L^t_n=D^t-W^t$ imamo
$$
L^t_nf(x)=f(x)\sum_je^{-|x-x_j|^2/4t} - \sum_jf(x_j)e^{-|x-x_j|^2/4t},
$$

gdje $x\in V=\{x_1,...,x_n\}$ i $f\in L^2(M)$. Naravno, $L^t_n$ ima prirodno proširenje na prostor funkcija $f:M \rightarrow \mathbb{R}$ dan istim izrazom $L^t_nf(p)=f(p)\sum_je^{-|p-x_j|^2/4t} - \sum_jf(x_j)e^{-|p-x_j|^2/4t}$, gdje $p\in M$.

Kako su $x_1,...,x_n$ uzorkovani po uniformnoj distribuciji $d\mu=\frac{dV}{vol(M)},$
vidimo da je
$$
\frac{1}{n}L^t_nf(p)=\frac{1}{n}\left(\sum_je^{-|p-x_j|^2/4t}f(p) - \sum_jf(x_j)e^{-|p-x_j|^2/4t}\right)$$
empirijsko usrednjenje očekivane vrijednosti
$$
 \mathbb{E}L^t_nf(p)=f(p)\int_Me^{-|p-y|^2/4t}d\mu-\int_Me^{-|p-y|^2/4t}f(y)d\mu
$$

Pokazali smo da u limesu $t\to 0$ vrijedi $\frac{1}{t(4\pi t)^{d/2}}\mathbb{E}L^t_nf(p)\to \frac{\mathcal{L}f(p)}{vol(M)}$. Sada želimo odabrati takav niz $t_n$ da (u vjerojatnosti) $\frac{1}{t_n(4\pi t)^{d/2}}\frac{1}{n}L^{t_n}_nf(p)\to \frac{\mathcal{L}f(p)}{vol(M)}$ dok $n\to \infty$. Jasno je da $t_n$ treba težiti u $0$ da bi ovo vrijedilo, ali vidjet ćemo da ne smije prebrzo težiti u $0$ (želimo li primjeniti određene nejednakosti iz statistike).

**TEOREM** (*Hoeffding*)

Neka je $K > 0$. Pretpostavimo da su slučajne varijable $X_1,...,X_n$ (na prostoru vjerojatnosti $\Omega$ sa mjerom vjerojatnosti $\mathbb{P}$) neovisne i jednako distribuirane (tako da sve $X_i$ imaju iste očekivane vrijednosti $\mathbb{E}[X_i]$). Nadalje, pretpostavimo da $X_i$ gotovo sigurno zadovoljavaju $|X_i|\leq K$, tj. $\mathbb{P}(X_i\in [-K,K])=1$, onda za svaki $\varepsilon >0$ imamo:
$$
\mathbb{P}(\frac{1}{n}\sum_iX_i-\mathbb{E}[X_i]>\varepsilon ) < \exp(-\frac{\varepsilon^2 n}{2K^2})
$$


Primijetimo da za dovoljno male $t$, $|e^{-|p-x_i|^2/4t}f(p)-e^{|p-x_i|^2/4t}f(x_i)|\leq 1$ (jer onda imamo $e$ na neki veliki negativni broj). Ograničimo se dakle na one $t$ za koje ovo vrijedi (ionako promatramo limes $t\to 0$).


Onda iz Hoeffdingove nejednakosti imamo:
$$
\mathbb{P}\left[\frac{1}{t}\frac{1}{(4\pi t)^{d/2}}\left( \frac{1}{n}L^t_nf(p)-\mathbb{E}L^t_nf(p)  \right)> \varepsilon\right] \leq \exp(-\varepsilon^2 n(4\pi t)^{d+2}).
$$

Dakle, vidimo da odabirom:

$$
t_n=\left(\frac{1}{n}\right)^{\frac{1}{d+2+\alpha}}
$$
za bilo koji $\alpha>0$, desna strana teži u $0$. Primijetimo doduše da npr. za $t_n=\left( \frac{1}{n}\right)^{\frac{1}{d+2}}$ nemamo istu situaciju unatoč tome što je tako potencija veća pa $t$ brže trne, tj. čak i ako dobijemo željeni rezultat i u tom slučaju, Hoeffdingova jednakost nije dovoljno jaka da to garantira.

Ovim smo pokazali da nenormalizirani laplasijan grafa (u vjerojatnosti) teži u $\frac{\mathcal{L}f(p)}{vol(M)}$, ali to u stvari znači da normalizirani laplasijan teži u Laplace-Beltramijev operator $\mathcal{L}$.

Kako $t\to 0$, imamo $(4\pi t)^{-d/2}\int e^{-|p-y|^2/4t}d\mu \to \frac{1}{vol(M)}$ (jer heat kernel teži u Diracov delta pa integral po mjeri $dV$ iznosi $1$, tj. po mjeri $d\mu$ je $1/vol(M)$). Očito je $\frac{1}{n}\sum_i e^{-|p-x_i|^2/4t}$ empirijsko usrednjenje očekivane vrijednosti $\int e^{-|p-y|^2/4t} d\mu$.

Sada iz $\frac{1}{t_n(4\pi t)^{d/2}}\frac{1}{n}L_nf(p)\to \frac{\mathcal{L}f(p)}{vol(M)}$ vidimo da i
$$\frac{1}{t_n}D_n^{-1}L_nf(p)=\frac{1}{t_n}((4\pi t_n)^{d/2} n D_n^{-1})\frac{1}{t_n(4\pi t)^{d/2}}\frac{1}{n}\mathcal{L}f(p)\to \mathcal{L}f(p).$$

Naime, kako je $(D_n)_{ii}=\sum_je^{-|x_i-x_j|^2/4t_n}$ dijagonalna, množenje sa $(4\pi t_n)^{d/2} n D_n^{-1}$ odgovara dijeljenju sa $(4\pi t_n)^{-d/2}\frac{1}{n}\sum_i e^{-|p-x_i|^2/4t_n}$ (gdje smo naravno uzeli proširenje na funkcije $f:M\rightarrow \mathbb{R}$), što u limesu prelazi u dijeljenje s $1/vol(M)$ pa je rezultat limesa samo $\mathcal{L}f(p)$.


Zapravo, ovdje smo samo pokazali da odgovarajući operatori konvergiraju (i tako motivirali odabir težina za algoritam). Prava konzistentnost mora garantirati da svojstvene funkcije konvergiraju, što je (opet za uniformnu distribuciju podataka) pokazano u *Convergence of Laplacian Eigenmaps*, Belkin and Niyogi.

## Algoritam

Sada je opis algoritma jednostavan, a  algoritam provodimo u 3 koraka:

Input: $x_1,...,x_n\in \mathbb{R}^D$, $\quad d$, $\quad k\in \mathbb{N}$

Za input uzimamo i $t$, iako smo gore pokazali kako odabrati niz $t_n$ tako da se, striktno govoreći, ta varijabla može izbjeći.

1.   Definirajmo težine \begin{equation*}
    w_{ij}=
    \begin{cases}
      e^{-|x_i-x_j|^2/4t}, & \text{ako su}\ x_i\ \text{i}\ x_j\ \text{među}\ k\ \text{najbližih susjeda} \\
      0, & \text{inače}
    \end{cases}
  \end{equation*}
2.   Pronađimo svojstvene vektore $y$ normaliziranog laplasijana, tj. riješimo problem $Ly=\lambda Dy$. Vratimo matricu $U$ čiji su stupci svojstveni vektori.
3.   Vrati $Y=[U]_{n\times d}$, tj. matricu prvih $d$ svojstvenih vektora (normaliziranog) laplasijana grafa, odnosno izbaci posljednjih $n-d$ stupaca matrice $U$ dobivene u prethodnom koraku.




## Veza sa locally linear embedding algoritmom?

U svom radu (1) Belkin i Niyogi su pokazali da su  (pod određenim uvjetima) matrica $E=(I-W)^T(I-W)$ i kvadrat Laplace-Beltramijevog operatora $\mathcal{L}^2$ približno proporcionalni. Kako $\mathcal{L}^2$ i $\mathcal{L}$ imaju iste svojstvene vektore, iz ovoga (pod tim uvjetima) odmah slijedi ekvivalentnost LLE i LE algoritama.

## Dodatak: Hoeffdingov teorem

**LEMA** (Hoeffding)

Neka je X (realna) slučajna varijabla sa očekivanom vrijednosti $\mathbb{E}[X]=\eta$ koja gotovo sigurno upada u interval $[a,b]$, odnosno $\mathbb{P}(X\in [a,b])=1$. Onda za sve $\lambda\in \mathbb{R}$ imamo
$$
\mathbb{E}[e^{\lambda X}]\leq \exp(\lambda \eta+ \frac{\lambda^2(b-a)^2}{8})]
$$

*DOKAZ*




*   Možemo pretpostaviti da $\mathbb{E}[X]=0$. Naime, varijablu $X$ zamijenimo s $X-\mathbb{E}[X]$ pa ako pokažemo $\mathbb{E}[e^{\lambda(X-\mathbb{E}[X])}]\leq \exp(\frac{\lambda^2(b-a)^2}{8})$, onda automatski vrijedi i $\mathbb{E}[e^{\lambda X}]e^{-\lambda \mathbb{E}[X]}\leq \exp(\frac{\lambda^2(a-b)^2}{8})$, što daje opći rezultat.

*   Prisjetimo se da za konveksnu funkciju $f$ vrijedi $f((1-t)a+tb)\leq (1-t)f(a)+tf(b)$ za sve $t\in[0,1]$. Drugim riječima, graf konveksne funkcije na segmentu od $a$ do $b$ mora ležati ispod pravca koji veže $(a,f(a))$ i $(b,f(b))$. Nadalje, primijetimo da za svaki $x\in[a,b]$ vrijedi $x=\frac{b-x}{b-a}a+\frac{x-a}{b-a}b$ i da možemo uzeti $\frac{b-x}{b-a}=(1-t)\in[0,1]$ te $\frac{x-a}{b-a}=t\in[0,1]$.


*   Kako je $x\mapsto e^{\lambda x}$ konveksna, vrijedi $e^{\lambda x}\leq \frac{b-x}{b-a}e^{\lambda a}+\frac{x-a}{b-a}e^{\lambda b}$ za sve $x\in[a,b]$. Iz ovoga vidimo da očekivana vrijednost zadovoljava:
$$
\mathbb{E}[e^{\lambda X}]\leq \frac{b-\mathbb{E}[X]}{b-a}e^{\lambda a}+\frac{\mathbb{E}[X]-a}{b-a}e^{\lambda b}=\frac{b}{b-a}e^{\lambda a}-\frac{a}{b-a}e^{\lambda b}
$$
jer $\mathbb{E}[X]=0$.

*   Definirajmo $u=\lambda (b-a)$ i $\theta =-\frac{a}{b-a}>0$ (iz $\mathbb{E}[X]=0$ slijedi $a\leq 0 \leq b$). Nadalje, neka je $\varphi(u)=-\theta u +\log(1-\theta +\theta e^u)$ tako da
$$\frac{b}{b-a}e^{\lambda a}-\frac{a}{b-a}e^{\lambda b}=(1-\theta)e^{\lambda a}+\theta e^{\lambda b}=(1-\theta + \theta e^{\lambda (b-a)})e^{-\lambda \theta(b-a)}=(1-\theta+\theta e^u)e^{-\theta u}=e^{\varphi(u)}.$$
Dakle, $\mathbb{E}[e^{\lambda X}]\leq e^{\varphi(u)}$ i još je jedino potrebno omeđiti samu funkciju $\varphi$.

*   Po Taylorovom teoremu $\varphi(u)=\varphi(0)+u\varphi'(u)+\frac{1}{2}u^2\varphi''(v)$ za neki $v\in [0,u]$. Sada primijetimo da $\varphi(0)=0, \quad \varphi'(u)=-\theta + \frac{\theta e^u}{1-\theta+\theta e^u}\vert_{u=0}=0$ te
$$
\varphi''(v)=\frac{\theta e^v (1- \theta + \theta e^v)-\theta^2 e^{2v}}{(1-\theta+\theta e^v)^2}=\frac{\theta}{1-\theta+\theta e^v}\left(1- \frac{\theta e^v}{1-\theta + \theta e^v} \right)=t(1-t),
$$
za $t=\frac{\theta e^v}{1- \theta+ \theta e^v}$.


*   Kako parabola $t-t^2$ ima maksimum u $x=\frac{1}{2}$, gdje iznosi $\frac{1}{4}$, imamo $\varphi''(v)\leq \frac{1}{4}$. Ovo sve nam konačno daje
$$
\varphi(u)\leq 0+u\cdot0+\frac{1}{2}u^2\frac{1}{4}=\frac{1}{8}\lambda^2(b-a)^2,
$$
iz čega tvrdnja teorema odmah slijedi

**TEOREM** (*Hoeffding*)

Pretpostavimo da svaka od neovisnih slučajnih varijabli $X_1,...,X_n$ (na prostoru vjerojatnosti $\Omega$ s mjerom vjerojatnosti $\mathbb{P}$) gotovo sigurno upada u svoj interval $[a_i,b_i]$, tj. $\mathbb{P}(X_i\in [a_i,b_i])=1$. U tom slučaju $S_n=X_1+\dots +X_n$ zadovoljava sljedeću među
$$
\mathbb{P}(S_n-\mathbb{E}(S_n)\geq t)\leq \exp(-\frac{2t^2}{\sum_i (b_i-a_i)^2}).
$$




---



*Commentarii*

*   U posebnoj situaciji gdje su sve slučajne varijable jednako distribuirane i sve gotovo sigurno upadaju u isti $[a,b]$ imamo
$$\mathbb{P}(\frac{1}{n}S_n-\frac{1}{n}\mathbb{E}[S_n]\geq t)=\mathbb{P}(S_n-\mathbb{E}[S_n]\geq nt)\leq \exp \left( -\frac{2t^2n^2}{\sum_i(b-a)^n}\right)=\exp\left( -\frac{2t^2n^2}{n(b-a)^2} \right)=\exp\left( -\frac{2t^2n}{(b-a)^2} \right)$$
Kako su $X_i$ jednako distribuirane, imamo $\mathbb{E}[X_i]=\eta$ za sve $i$, što daje $\frac{1}{n}\mathbb{E}[S_n]=\frac{1}{n}\mathbb{E}[X_1+\dots x_n]=\frac{1}{n}(n\eta)=\eta$, stoga $\mathbb{P}(\overline{X} - \eta \geq t)\leq e^{-\frac{2t^2n}{(b-a)^2}}$. Ovdje je $\overline{X}=\frac{1}{n}(X_1+\dots +X_n)$ srednja vrijednost uzorka $X_1,...,X_n$.

* Štoviše, ako je interval simetričan $[a,b]=[-K,K]$, tj. $|X_i|\leq K$ gotovo sigurno, onda $(b-a)^2=4K^2$ pa imamo $\mathbb{P}(\overline{X} - \eta \geq t)\leq e^{-\frac{2t^2n}{4K^2}}=e^{-\frac{t^2n}{2K^2}}$.



---



*DOKAZ*

Koristeći Markovljevu nejednakost $\mathbb{P}(X\geq a) \leq \frac{\mathbb{E}[X]}{a}$, imamo (za sve $s,t>0$)
$$
\mathbb{P}(S_n-\mathbb{E}[S_n]\geq t)= \mathbb{P}(e^{s(S_n-\mathbb{E}[S_n])}\geq e^{st})\leq e^{-st}\mathbb{E}[e^{s(S_n-\mathbb{E}[S_n])}],
$$
a kako su $X_i$ neovisne, imamo dalje:
$$
=e^{-st}\Pi_{i=1}^n\mathbb{E}[e^{s(X_i-E[X_i])}]\leq e^{-st}\Pi_{i=1}^n e^{s^2(b_i-a_i)^2/8}=\exp \left( -st+ \frac{1}{8}s^2\sum_i(b_i-a_i)^2 \right),
$$
gdje nejednakost slijedi iz Hoeffdingove leme.

Ovaj izraz poprima minimum upravo tamo gdje je $g(s)=-st+ \frac{1}{8}s^2\sum_i(b_i-a_i)^2$ minimalan. $g$ je kvadratna funkcija oblika $ax^2+bx$ (a>0) koja minimum mora poprimiti u $0=g'(x)=2ax+b\implies x=-\frac{b}{2a}$, a u toj točki iznosi $g(x)=-\frac{b^2}{4a}$. U našem slučaju $b=-t$ i $a=\frac{1}{8}\sum_i(b_i-a_i)^2$ pa $g$ poprima minimum u $s=\frac{4t}{\sum_i(b_i-a_i)^2}$, gdje iznosi $-\frac{2t^2}{\sum_i(b_i-a_i)^2}$.

Supstitucijom minimalne vrijednosti u prethodno dobivenu nejednakost (koja vrijedi za sve $s>0$):
$$\mathbb{P}(S_n-\mathbb{E}[S_n]\geq t)\leq e^{g(s)},$$
slijedi tvrdnja teorema.

**LITERATURA**
1. Diffusion maps, Coifman and Lafon

Laplacian eigenmaps algoritam se može interpretirati stohastički. Ovo dovodi do poboljšanja LE algoritma i garancije konzistentnosti čak i u slučaju neuniformne distribucije.

## Markovljev kernel

Neka je $\mathbb{P}: \Sigma \rightarrow [0,1]$ mjera vjerojatnosti na $\mathbb{R}^k$ koja uzima događaje $A\subset \mathbb{R}^k$ iz neke kolekcije događaja (tj. Borelove $\sigma$-algebre) $\Sigma$ i daje broj $\mathbb{P}(A)\in [0,1]$. Za funkciju $p:\mathbb{R}^k\times \Sigma\rightarrow [0,1]$ kažemo da je **Markovljev kernel** na $\mathbb{R}^k$ ako je $p(x,dx)=p(x,\cdot):\Sigma \rightarrow [0,1]$ mjera vjerojatnosti za svaki $x\in \mathbb{R}^k$ i ako je $p(\cdot,A):\mathbb{R}^k\rightarrow[0,1]$ izmjerljiva (pa se može integrirati). Dakle, pomoću $p$ smo parametrizirali mjere vjerojatnosti točkama iz $\mathbb{R}^k$.

Za niz slučajnih varijabli $X_1,X_2,...$ (tj. stohastički proces) ćemo kazati da je **Markovljev proces** (ili Markovljev lanac) ako vjerojatnost za prelazak iz jednog stanja u sljedeće ovisi samo trenutnom stanju, a ne o prethodnim stanjima. Preciznije: $\mathbb{P}(X_{n+1}=x_{n+1} \vert X_1=x_1,...,X_n=x_n)=\mathbb{P}(X_{n+1}=x_{n+1} \vert X_n=x_{n})$. Kažemo da je Markovljev proces vremenski homogen ako $\mathbb{P}(X_{n+1}=x \vert X_n=y)=\mathbb{P}(X_{n}=x\vert X_{n-1}=y)$.


Neka je $X_n$ vremenski homogen Markovljev proces. Primijetimo da je $p_n(x,A)=\mathbb{P}(X_n\in A \vert X_0=x)$ Markovljev kernel kojeg zovemo **tranzicijskim kernelom**. Tranzicijski kernel nam daje vjerojatnost da ćemo se u $n$-tom koraku naći u skupu $A$ ako smo krenuli iz $x$. Iz vremenske homogenosti vidimo da $\mathbb{P}(X_n\in A \vert X_{n-1}=x)=\mathbb{P}(X_1\in A \vert X_0=x)=p_1(x,A)$ pa $p_1$ predstavlja vjerojatnost prijelaza u svakom koraku.



Neka je $X_n$ vremenski homogen Markovljev proces s tranzicijskim kernelom $p_n$, neka $p_1=p$ i neka je $X_0$ distribuiran po mjeri vjerojatnosti $\mu$. Pod tim pretpostavkama nije teško pokazati da vektorska slučajna varijabla $(X_0,...,X_n)$ mora imati distribuciju $d\mu(x_0)p(x_0,dx_1)p(x_1,dx_2)\dots p(x_{n-1},dx_n)$. Drugim riječima,

$$\mathbb{P}(X_0\in A_0,...,X_n\in A_n)=\int_{A_0}\dots \int_{A_{n-1}}p(y_{n-1}, A_n)p(y_{n-2}dy_{n-1})\dots p(y_0,dy_1)d\mu(y_0).$$

Naime,
$$
\mathbb{P}(X_0\in A, X_1\in B)=\int_A \mathbb{P}(X_1\in B \vert X_0=x)d\mu (x)= \int_A p(x,B)d\mu (x)\\=\int_A\int_B p(x,dy)d\mu (x) =\int_{A\times B} p(x,dy)d\mu (x)
$$
Dalje imamo i
$$
\mathbb{P}(X_0\in A, X_1\in B, X_2\in C)=\int_{A\times B} \mathbb{P}(X_2\in C\vert X_0=x, X_1=y)p(x,dy)d\mu (x)=\int_{A\times B} \mathbb{P}(X_2\in C\vert X_1=y)p(x,dy)d\mu (x) =\int_{A\times B} p(y,C) p(x,dy)d\mu (x)=\int_{A\times B \times C} p(y,dz) p(x,dy)d\mu (x),
$$
gdje druga jednakost slijedi iz činjenice da je $X_n$ Markovljev proces. Sada je jasno da možemo nastaviti induktivno.



Također, imamo i svojevrsni obrat. Ako je $p$ Markovljev kernel na $\mathbb{R}^k$, onda (po Kolmogorovljevom teoremu) možemo konstruirati vremenski homogeni Markovljev proces $X_n$ tako da je $(X_0,X_1,...,X_n)$ distribuiran po $d\mu(x_0)p(x_0,dx_1)p(x_1,dx_2)\dots p(x_{n-1},dx_n)$, gdje je $\mu$ neka početna distribucija (tj. distribucija od $X_0$).

Dakle, za konstrukciju cijelog procesa je dovoljno znati samo tranzicijski kernel i početnu distribuciju.

*   Kažemo da je $p$ reverzibilan s obzirom na mjeru $\mu$ ako $\int_A p(x,B)d\mu (x)=\int_B p(x,A) d\mu (x)$ (tj. $\mathbb{P}(X_1\in B, X_0\in A)= \mathbb{P}(X_1\in A, X_0\in B)$) jer onda $(X_{n+1},...,X_{n+k})$ i $(X_{n+k},...,x_{n+1})$ imaju istu distribuciju, odnosno $\mathbb{P}(X_0\in A_0,...,X_n\in A_n)=\mathbb{P}(X_0\in A_n,...,X_n\in A_0)$.

*   Posebno, ako je $p$ reverzibilan s obzirom na $\mu$, a $\Omega$ je ukupni prostor događaja (kojeg svaka mjera vjerojatnosti šalje u $1$), onda $\int_\Omega p(x,B)d\mu (x)=\int_B p(x,\Omega)d\mu (x)=\int_B d\mu=\mu(B)$. Dakle, $X_n$ je stacionaran s obzirom na $\mu$, odnosno ako $X_0$ ima distribuciju $\mu$, onda $\mathbb{P}(X_n\in A)=\mu (A)$ pa $X_n$ isto ima distribuciju $\mu$.

*   Neka je $p(x,y)$ gustoća mjere $p(x,dy)$ s obzirom na $\mu$ (tj. neka je Radon-Nikodymovljeva derivacija), onda $p(x,dy)=p(x,y)d\mu (y)$. $p(x,y)$ predstavlja vjerojatnost da će $X_1$ završiti na $y$ ako $X_0$ krene sa $x$, tj. da ćemo sa $x$ završiti na $y$. Ako je $p(x,y)d\mu (x)= p(y,x) d\mu (y)$, onda je $p(x,dy)$ reverzibilna s obzirom na $\mu$:
$$
\int_A p(x,B) d\mu (x)=\int_A \left( \int_B p(x,y)d\mu (y) \right) d\mu (x) = \int_{A\times B} p(x,y)d\mu (x) d\mu (y) =\int_{A\times B} p(y,x)d\mu (y) d\mu (x) = \int_B \left( \int_A p(y,x)d\mu (y) \right) d\mu (x)=\int_B p(x,A) d\mu (x)
$$

## Algoritam

Neka je $(X,\Sigma, \mu)$ prostor mjere. Dataset živi u $X$, a $\mu$ je raspodjela podataka. Neka je $k:X\times X\rightarrow \mathbb{R}$ (nenormalizirana) vjerojatnost da sa $x\in X$ dođemo na $y\in X$. Intuitivno, $k$ opisuje difuzijski proces. Pretpostavljamo da $k(x,y)=k(y,x)$, $k(x,y)\geq 0$ te da je $k$ integrabilna.


Mogući odabir je $k(x,y)=\exp(-\frac{|x-y|^2}{\varepsilon})$ za neki $\varepsilon>0$.

$k$ se može normalizirati pa zaista postane Markovljev kernel:

Neka je $d(x)=\int_X k(x,y)d\mu (y)$, onda $p(x,y)=\frac{k(x,y)}{d(x)}$ zadovoljava $\int_X p(x,y) d\mu(y)=1$. $p$ doduše više nije simetričan $p(x,y)\neq p(y,x)$, ali simetrija od $k$ garantira da je $p$ reverzibilan s obzirom na gustoću $\pi (y)=\frac{d(y)}{\int_X d(x) d\mu (x)}$ (tj. s obzirom na mjeru $\pi d\mu$) jer očito imamo $\pi (x)\frac{k(x,y)}{d(x)}=\pi (y)\frac{k(x,y)}{d(y)}$ (prema simetričnosti funkcije $k$ i definiciji $\pi$).

Pretpostavimo da je $k(x,y)=h(\frac{|x-y|^2}{\varepsilon})$ i $h$ eksponencijalno trne. Neka je $q$ gustoća od $\mu$, tj. $d\mu=qdx$, gdje je $dx$ Lebesgueova mjera (volumna forma) na $X=M$. Prije normalizacije $p(x,y)=\frac{k(x,y)}{d(x)}$ radimo još jednu normalizaciju:

Za $\alpha \in \mathbb{R}$ definirajmo $k^{(\alpha)}(x,y)=\frac{k(x,y)}{d^\alpha(x)d^\alpha(y)}$, gdje je $d^{\alpha}(x)=(d(x))^\alpha$ potenciranje s $\alpha$. Sada uzmemo $d^{(\alpha)}=\int_X k^{(\alpha)}(x,y)d\mu (y)$ i stavimo $p_{\varepsilon, \alpha}(x,y)=\frac{k^{(\alpha)}(x,y)}{d^{(\alpha)}(x)}$. Definiramo i operator $P_{\varepsilon  \alpha}f(x)=\int_Xp_{\varepsilon , \alpha}(x,y)f(y)d\mu (y)$.


**PROPOZICIJA**

Operator $L_{\varepsilon \alpha}=I- P_{\varepsilon \alpha}$ zadovoljava
$$
\lim_{\varepsilon \to 0}\frac{1}{\varepsilon}L_{\varepsilon \alpha}f = \frac{\mathcal{L} (fq^{1-\alpha})}{q^{1-\alpha}}- \frac{\mathcal{L} (q^{1-\alpha})}{q^{1-\alpha}}f.
$$
Dokaz je dan u Coifman i Lafon (1).

Neka je zadan dataset $x_1,...,x_n\in X$. Definiramo matricu $W$ sa unosima $w_{ij}=k(x_i,x_j)$ i $d(x_i)=\sum_jw_{ij}=\sum_jk(x_i,x_j)$ (što je empirijski oblik integrala $\int k(x,y)d\mu (y)$).

Dalje nastavljamo s normalizacijama kao i prije. $w^{(\alpha)}_{ij}=\frac{w{ij}}{d^\alpha(x_i)d^\alpha(x_j)}$, što se može zapisati i kao $W^{(\alpha)}=(D^{\alpha})^{-1}W(D^{\alpha})^{-1}$, gdje je $D$ dijagonalna matrica sa unosima $D_{ii}=d(x_i)=\sum_jw_{ij}$.



---



Naime, za dijagonalnu matricu
$A= \text{diag}(a_1,...,a_n)$ imamo $A^{-1}=\text{diag}(\frac{1}{a_1},...,\frac{1}{a_n})$ i $A^{\alpha}=\text{diag}(a_1^\alpha,...,a_n^\alpha)$

Također, za bilo koju matricu $B$ imamo $(ABA)_{ij}=e^iABAe_j=a_ie^iBe_ja_j=a_iB_{ij}a_j=A_{ii}B_{ij}A_{jj}$ jer za dijagonalnu $A$ imamo $Ae_j=a_je_j$, tj. $e^iA=a_ie^i$. Isto tako $(AB)_{ij}=A_{ii}B_{ij}$ i $(BA)_{ij}=A_{jj}B_{ij}$. Iz ovoga je sada očito da $W^{(\alpha)}=(D^{\alpha})^{-1}W(D^{\alpha})^{-1}$ za unose ima $w^{(\alpha)}_{ij}=\frac{w{ij}}{d^\alpha(x_i)d^\alpha(x_j)}$.



---



Konačno, stavimo $P=(D^{(\alpha)})^{-1}W^{(\alpha)}$ i definirajmo $L=I-P$. Dakle, $L$ je normalizirani laplasijan grafa koji za težine ima $w_{ij}^{(\alpha)}$. Primijetimo da smo prvo normalizirali težine, a onda uzeli normalizirani laplasijan.

Za $\alpha=0$ dobijemo samo laplasijan grafa s nenormaliziranim težinama $w_{ij}=k(x_i,x_j)$. Ako je distribucija podataka uniformna, onda zaista, kao što smo i pokazali, $\frac{1}{\varepsilon}Lf\to \frac{\mathcal{L}(fq)}{q}-\frac{\mathcal{L}q}{q}f=\mathcal{L}f$ jer je $q$ konstantan pa drugi član propada, a u prvom može izaći van diferencijalnog operatora.

Ako $q$ nije uniforman, onda za $\alpha=0$ ne dobijemo konvergenciju u Laplace-Beltramijev operator. Ipak, ako uzmemo težine sa normalizacijom $\alpha=1$, onda je jasno da *uvijek* imamo $\frac{1}{\varepsilon}L_{\varepsilon \alpha}\to \mathcal{L}f$.

Dakle, pravilno normalizirane težine nam garantiraju konzistentnost i za neuniformno uzorkovane podatke.

## Otvoreni problemi?

U svom radu Lafon i Coifman su pokazali da paravilno normalizirani laplasijan grafa teži u Laplace-Beltrami operator, ali nisu pokazali i konvergenciju svojstvenih vektora (rezultat koji su za uniformnu distribuciju dobili Niyogi i Belkin).

Ležernom čitatelju (kao što je autor) nije u potpunosti jasno slijedi li rezultat konvergencije svojstvenih vektora (za neuniformnu distribuciju) lagano kobiniranjem ta dva rada, odnosno je li već netko taj rezultat objavio, ili je problem još uvijek otvoren.
